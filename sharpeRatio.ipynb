# SHARPE RATION

```python
appl = pd.read_csv('apple.csv',index_col='Date', parse_dates=True)
```

```python
msft = pd.read_csv('msft.csv', index_col='Date', parse_dates=True)
```

```python
appl['daily return'] = appl['Adj Close'].pct_change(1)
```

```python
msft['daily return'] = msft['Adj Close'].pct_change(1)
```

```python
appl['daily return'].std()
```

```python
def compute_SR(df,risk_freerate=0):
    meanReturn = df['daily return'].mean()
    std = df['daily return'].std()
    
    sharpeRatio = (meanReturn-risk_freerate)/std
    
    return sharpeRatio * (252**0.5)
```

```python
sr_apple = compute_SR(appl)
sr_apple
```

```python
sr_msft = compute_SR(msft)
sr_msft
```
